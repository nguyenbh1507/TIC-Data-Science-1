## 3. Basic graph types

### 3.1 Line plot
- plot function graph
- plot time-series data

```python
# function graph
import numpy as np
# create data
x = np.arange(-1,3, 0.1)
y1 = x**2
y2 = np.exp(x)

plt.plot(x, y1, 'b-', label='Squared x') # plot 1st function
plt.plot(x, y2, 'r--', label='Exp(x)')   # plot 2nd function
plt.title('Function graph') # add title
plt.xlabel('x')             # add x label
plt.ylabel('y')             # add y label
plt.grid(True) # add grid
plt.legend()   # add legend
plt.show()     # show the graph
```

```python
# time-series data
# create data
import pandas as pd
df = pd.read_csv('weather.csv')

# create new figure, adjust its size
plt.figure(figsize=(10,5))
plt.plot(df.MinTemp, label='Minimum temperature')
plt.plot(df.MaxTemp, label='Maximum temperature')

plt.title('Temperature') # add title
plt.xlabel('Day')             # add x label
plt.ylabel('Temperatue, degree Celsius')             # add y label
plt.grid(True) # add grid
plt.legend()   # add legend
plt.show()     # show the graph
```

### 3.2 Scatter plot
Used to evaluate trends and correlation between to values

```python
# read the data
df1 = pd.read_csv('telcom.csv')
df1.MonthlyCharges.fillna(df1.MonthlyCharges.median(), inplace=True)

# plot MonthlyCharges vs TotalCharges
plt.scatter(x=df1.MonthlyCharges, y=df1.TotalCharges)
plt.xlabel('Monthly Charges')
plt.ylabel('Total Charges')
plt.show()
```

### 3.3 Histogram
- Show the distribution
- Detect outliers

```python
# Minimum temperature distribution
plt.hist(df.MinTemp, bins=15, density=True, label='Experiment') # if density=False/None, counts will be shown
plt.title('Temperature distribution')
plt.xlabel('Minimum temperature')
plt.ylabel('Density')

# # advanced part - add kde - Kernel Density Estimates
# from scipy import stats
# kde = stats.gaussian_kde(df.MinTemp)
# xx = np.arange(df.MinTemp.min(), df.MinTemp.max()+0.1, 0.1)
# plt.plot(xx,kde(xx), label='Estimated')

plt.legend()
plt.show()
```

### 3.4 Pie plot

```python
churn = df1.Churn.value_counts()
labels = churn.keys()
values = churn.values

# Pie graph of churn and non-churn customer
plt.pie(x=values, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.title('Churn customer distribution')
plt.show()
```

### 3.5 Bar plot
Alternative line

```python
# plot tenure vs customerID
plt.bar(df1.index, df1.tenure)
plt.xticks(df1.index, df1.customerID, rotation='vertical')
plt.xlabel('Customer ID')
plt.ylabel('Tenure')
plt.show()
```
